&nbsp;&nbsp;&nbsp;&nbsp; Любой бизнес хочет максимизировать количество клиентов. Для достижения этой цели важно не только пытаться привлечь новых, но и удерживать уже существующих. Удержать клиента обойдется компании дешевле, чем привлечь нового. Кроме того, новый клиент может оказаться слабо заинтересованным в услугах бизнеса и с ним будет сложно работать, тогда как о старых клиентах уже есть необходимые данные по взаимодействию с сервисом.
 <br>

&nbsp;&nbsp;&nbsp;&nbsp; Соответственно, прогнозируя отток, мы можем вовремя среагировать и попытаться удержать клиента, который хочет уйти. Опираясь на данные об услугах, которыми пользуется клиент, мы можем сделать ему специальное предложение, пытаясь изменить его решение об уходе от оператора. Благодаря этому задача удержания будет легче в реализации, чем задача привлечения новых пользователей, о которых мы еще ничего не знаем.<br>

&nbsp;&nbsp;&nbsp;&nbsp; Вам предоставлен набор данных от телекоммуникационной компании. В данных содержится информация о почти шести тысячах пользователей, их демографических характеристиках, услугах, которыми они пользуются, длительности пользования услугами оператора, методе оплаты, размере оплаты.
<br>

&nbsp;&nbsp;&nbsp;&nbsp; Cтоит задача проанализировать данные и спрогнозировать отток пользователей (выявить людей, которые продлят контракт и которые не продлят). Работа должна включать в себя следующие обязательные пункты:
1. Описание данных *(с расчетом базовых статистик)*,
2. Исследование зависимостей и формулирование гипотез,
3. Построение моделей для прогнозирования оттока *(с обоснованием выбора той или иной модели)* на основе проверенных гипотез и выявленных взаимосвязей,
4. Сравнение качества полученных моделей.

# 1. Описание данных

```python
import pandas as pd
df_data = pd.read_csv('telecom_users.csv')
# Проверка информации о данных
print(df_data.info())
```

```python
# Проверка на некорректные значения в 'TotalCharges'
invalid_values_mask = df_data['TotalCharges'].astype(str).str.match(r'^[+-]?(\d+(\.\d*)?|\.\d+)$') == False

# Отображаем некорректные значения
invalid_values = df_data['TotalCharges'][invalid_values_mask]

# Вывод некорректных значений
print("Некорректные значения в 'TotalCharges':")
print(invalid_values.unique())

# Подсчет количества некорректных записей
count_invalid_values = invalid_values_mask.sum()
print(f"Количество некорректных записей в 'TotalCharges': {count_invalid_values}")

# Преобразуем 'TotalCharges' в числовой формат, заменяя некорректные значения на NaN
df_data['TotalCharges'] = pd.to_numeric(df_data['TotalCharges'], errors='coerce')

# Заменяем NaN (если они остались) на 0
df_data['TotalCharges'] = df_data['TotalCharges'].fillna(0)

# Преобразование 'TotalCharges' в тип float
df_data['TotalCharges'] = df_data['TotalCharges'].astype(float)

# Проверка результата
print(f"Количество некорректных записей в 'TotalCharges' после преобразования: {df_data['TotalCharges'].isna().sum()}")
```

```python
# Преобразование категориальных переменных
df_data['gender'] = df_data['gender'].map({'Male': 0, 'Female': 1})
df_data['Partner'] = df_data['Partner'].map({'Yes': 1, 'No': 0})
df_data['Dependents'] = df_data['Dependents'].map({'Yes': 1, 'No': 0})
df_data['PhoneService'] = df_data['PhoneService'].map({'Yes': 1, 'No': 0})
df_data['MultipleLines'] = df_data['MultipleLines'].map({'Yes': 1, 'No': 0, 'No phone service': 0})
df_data['InternetService'] = df_data['InternetService'].map({'DSL': 0, 'Fiber optic': 1, 'No': 2})
df_data['OnlineSecurity'] = df_data['OnlineSecurity'].map({'Yes': 1, 'No': 0, 'No internet service': 0})
df_data['OnlineBackup'] = df_data['OnlineBackup'].map({'Yes': 1, 'No': 0, 'No internet service': 0})
df_data['DeviceProtection'] = df_data['DeviceProtection'].map({'Yes': 1, 'No': 0, 'No internet service': 0})
df_data['TechSupport'] = df_data['TechSupport'].map({'Yes': 1, 'No': 0, 'No internet service': 0})
df_data['StreamingTV'] = df_data['StreamingTV'].map({'Yes': 1, 'No': 0, 'No internet service': 0})
df_data['StreamingMovies'] = df_data['StreamingMovies'].map({'Yes': 1, 'No': 0, 'No internet service': 0})
df_data['Contract'] = df_data['Contract'].map({'Month-to-month': 0, 'One year': 1, 'Two year': 2})
df_data['PaperlessBilling'] = df_data['PaperlessBilling'].map({'Yes': 1, 'No': 0})
df_data['PaymentMethod'] = df_data['PaymentMethod'].map({'Credit card (automatic)': 0,'Bank transfer (automatic)': 1,
                                                         'Mailed check': 2, 'Electronic check': 3})
df_data['Churn'] = df_data['Churn'].map({'Yes': 1, 'No': 0})

# Проверка типов данных
print(df_data.dtypes)

# Проверка на наличие пропущенных значений в DataFrame
missing_values = df_data.isnull().sum()

print("Количество пропущенных значений в каждом столбце:")
if missing_values.sum() == 0:
    print("Нет пропущенных значений в данных.")
else:
    print(missing_values[missing_values > 0])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Создание подграфиков
fig, axes = plt.subplots(nrows=6, ncols=3, figsize=(18, 30))
fig.suptitle('Анализ оттока клиентов по различным признакам', fontsize=20)

# Функция для построения графиков
def plot_churn_by_feature(df, feature, ax, title, xlabel, ylabel):
    churn_by_feature = df.groupby(feature)['Churn'].value_counts(normalize=True).unstack() * 100
    churn_by_feature.plot(kind='bar', stacked=True, ax=ax)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend(title='Отток', labels=['Остались', 'Покинули'])

# Графики для категориальных переменных
plot_churn_by_feature(df_data, 'Contract', axes[0, 0], 'Отток по типу контракта', 'Тип контракта (Month-to-month - 0, One year - 1, Two year - 2)', 'Процент клиентов')
plot_churn_by_feature(df_data, 'gender', axes[0, 1], 'Отток по полу', 'Пол (0 - Male, 1 - Female)', 'Процент клиентов')
plot_churn_by_feature(df_data, 'Dependents', axes[0, 2], 'Отток по наличию иждивенцев', 'Иждивенцы (0 - Нет, 1 - Да)', 'Процент клиентов')
plot_churn_by_feature(df_data, 'Partner', axes[1, 0], 'Отток по наличию партнера', 'Партнер (0 - Нет, 1 - Да)', 'Процент клиентов')
plot_churn_by_feature(df_data, 'InternetService', axes[1, 1], 'Отток по типу интернет-сервиса', 'Тип интернет-сервиса (0 - DSL, 1 - Fiber optic, 2 - Нет)', 'Процент клиентов')
plot_churn_by_feature(df_data, 'TechSupport', axes[1, 2], 'Отток по наличию технической поддержки', 'Техническая поддержка (0 - Нет, 1 - Да)', 'Процент клиентов')

# Графики для бинарных переменных
plot_churn_by_feature(df_data, 'OnlineSecurity', axes[2, 0], 'Отток по онлайн-безопасности', 'Онлайн-безопасность (0 - Нет, 1 - Да)', 'Процент клиентов')
plot_churn_by_feature(df_data, 'DeviceProtection', axes[2, 1], 'Отток по страховке оборудования', 'Страховка оборудования (0 - Нет, 1 - Да)', 'Процент клиентов')
plot_churn_by_feature(df_data, 'OnlineBackup', axes[2, 2], 'Отток по онлайн-резервированию', 'Онлайн-резервирование (0 - Нет, 1 - Да)', 'Процент клиентов')
plot_churn_by_feature(df_data, 'StreamingTV', axes[3, 0], 'Отток по стриминговому телевидению', 'Стриминговое телевидение (0 - Нет, 1 - Да)', 'Процент клиентов')
plot_churn_by_feature(df_data, 'StreamingMovies', axes[3, 1], 'Отток по стриминговому кинотеатру', 'Стриминговый кинотеатр (0 - Нет, 1 - Да)', 'Процент клиентов')
plot_churn_by_feature(df_data, 'PaperlessBilling', axes[3, 2], 'Отток по безбумажному биллингу', 'Безбумажный биллинг (0 - Нет, 1 - Да)', 'Процент клиентов')

# Создание категориальных переменных для MonthlyCharges и TotalCharges
df_data['MonthlyCharges_bins'] = pd.cut(df_data['MonthlyCharges'], bins=5, labels=[f'Range {i+1}' for i in range(5)])
df_data['TotalCharges_bins'] = pd.cut(pd.to_numeric(df_data['TotalCharges'], errors='coerce'), bins=5, labels=[f'Range {i+1}' for i in range(5)])
df_data['tenure_bins'] = pd.cut(df_data['tenure'], bins=5, labels=[f'Range {i+1}' for i in range(5)])

# Графики для разбитых на диапазоны переменных
plot_churn_by_feature(df_data, 'MonthlyCharges_bins', axes[4, 0], 'Отток по месячному размеру оплаты', 'Диапазоны месячного размера оплаты', 'Процент клиентов')
plot_churn_by_feature(df_data, 'TotalCharges_bins', axes[4, 1], 'Отток по общей сумме оплат', 'Диапазоны общей суммы оплат', 'Процент клиентов')
plot_churn_by_feature(df_data, 'tenure_bins', axes[4, 2], 'Отток по сроку обслуживания', 'Диапазоны срока обслуживания (месяцы)', 'Процент клиентов')

# Распределение оттока по методу оплаты
plot_churn_by_feature(df_data, 'PaymentMethod', axes[5, 0], 'Отток по методу оплаты', 'Метод оплаты\n(0 - Credit card (automatic), 1 - Bank transfer (automatic), 2 - Mailed check, 3 - Electronic check)', 'Процент клиентов')

# Удаляем пустые подграфики, если есть
for ax in axes.flat:
    if not ax.has_data():
        ax.remove()

plt.tight_layout(rect=[0, 0, 1, 0.96])  # Увеличиваем отступ сверху для заголовка
plt.show()
```

Диаграмма рассеяния для переменных MonthlyCharges_bins, TotalCharges_bins и tenure_bins с переменной Churn.
Заметная значимость по шкале Чеддока - при коэффициенте корреляции от 0.5 по модулю и выше.

```python
# Преобразование категориальных переменных в числовые
df_data['MonthlyCharges_bins'] = pd.cut(df_data['MonthlyCharges'], bins=5, labels=[f'Range {i+1}' for i in range(5)])
df_data['TotalCharges_bins'] = pd.cut(df_data['TotalCharges'], bins=5, labels=[f'Range {i+1}' for i in range(5)])
df_data['tenure_bins'] = pd.cut(df_data['tenure'], bins=5, labels=[f'Range {i+1}' for i in range(5)])

# Преобразуем бины в числовой формат для вычисления корреляции
df_data['MonthlyCharges_bins'] = df_data['MonthlyCharges_bins'].cat.codes
df_data['TotalCharges_bins'] = df_data['TotalCharges_bins'].cat.codes
df_data['tenure_bins'] = df_data['tenure_bins'].cat.codes

# Вычисление коэффициентов корреляции
correlation_monthly = df_data['Churn'].corr(df_data['MonthlyCharges_bins'])
correlation_total = df_data['Churn'].corr(df_data['TotalCharges_bins'])
correlation_tenure = df_data['Churn'].corr(df_data['tenure_bins'])

# Вывод результатов
print(f"Коэффициент корреляции между 'Churn' и 'MonthlyCharges_bins': {correlation_monthly:.2f} - {'больше' if abs(correlation_monthly) > 0.5 else 'меньше'} 0.5")
print(f"Коэффициент корреляции между 'Churn' и 'TotalCharges_bins': {correlation_total:.2f} - {'больше' if abs(correlation_total) > 0.5 else 'меньше'} 0.5")
print(f"Коэффициент корреляции между 'Churn' и 'tenure_bins': {correlation_tenure:.2f} - {'больше' if abs(correlation_tenure) > 0.5 else 'меньше'} 0.5")

# Проверка значимости и построение диаграмм рассеяния
significant_features = []

if abs(correlation_monthly) >= 0.5:
    significant_features.append('MonthlyCharges_bins')
    
if abs(correlation_total) >= 0.5:
    significant_features.append('TotalCharges_bins')

if abs(correlation_tenure) >= 0.5:
    significant_features.append('tenure_bins')

# Построение диаграмм рассеяния для значимых признаков
for feature in significant_features:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=df_data[feature], y=df_data['Churn'], alpha=0.6)
    plt.title(f"Диаграмма рассеяния: {feature} vs Churn")
    plt.xlabel(feature)
    plt.ylabel("Отток (Churn)")
    plt.axhline(y=0.5, color='r', linestyle='--', label='Порог оттока')
    plt.legend()
    plt.show()
```

```python
# Выбор переменных для анализа
variables = ['MonthlyCharges', 'TotalCharges', 'tenure']

# Создание подграфиков для ящиков с усами
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 18))
fig.suptitle('Ящики с усами для переменных', fontsize=20)

for i, column in enumerate(variables):
    # Ящик с усами для текущей переменной
    sns.boxplot(x='Churn', y=column, data=df_data, ax=axes[i])
    axes[i].set_title(f'{column} в зависимости от оттока')
    axes[i].set_xlabel('Отток (0 - Остались, 1 - Покинули)')
    axes[i].set_ylabel(column)

    # Анализ выбросов
    Q1 = df_data[column].quantile(0.25)
    Q3 = df_data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Проверка наличия выбросов
    num_outliers = ((df_data[column] < lower_bound) | (df_data[column] > upper_bound)).sum()
    if num_outliers > 0:
        outlier_text = f"Наличие выбросов: {num_outliers} выбросов"
    else:
        outlier_text = "Наличие выбросов: отсутствуют"

    # Добавление текста под графиком
    axes[i].text(0.5, -0.15, outlier_text, ha='center', va='top', transform=axes[i].transAxes)

plt.tight_layout(rect=[0, 0, 1, 0.96])  # Увеличиваем отступ сверху для заголовка
plt.show()
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

def find_optimal_clusters(data, max_clusters=10):
    # Проверка на наличие числовых данных
    if data.select_dtypes(include=[np.number]).shape[1] == 0:
        print("Нет числовых данных для кластеризации.")
        return
    
    # Нормализация данных
    scaler = StandardScaler()
    data_scaled = scaler.fit_transform(data)

    # Метод локтя для определения оптимального количества кластеров
    wcss = []  # within-cluster sum of squares
    for i in range(1, max_clusters + 1):
        kmeans = KMeans(n_clusters=i, random_state=42)
        kmeans.fit(data_scaled)
        wcss.append(kmeans.inertia_)  # Сохраняем значение WCSS

    # Визуализация метода локтя
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, max_clusters + 1), wcss, marker='o')
    plt.title('Метод локтя для определения оптимального количества кластеров')
    plt.xlabel('Количество кластеров')
    plt.ylabel('WCSS')
    plt.grid()
    plt.show()

# Обновленные наборы признаков
features_sets = {
    'Set 1': df_churn[['MonthlyCharges', 'tenure']],
    'Set 2': df_churn[['MonthlyCharges', 'TotalCharges']],
    'Set 3': df_churn[['tenure', 'TotalCharges']],
    'Set 4': df_churn[['MonthlyCharges'] + [col for col in df_churn.columns if 'InternetService_' in col]]
}

# Для каждого набора признаков находим оптимальное количество кластеров
for name, feature_set in features_sets.items():
    print(f'Оптимальное количество кластеров для {name}:')
    find_optimal_clusters(feature_set.select_dtypes(include=[np.number]))  # Используем только числовые столбцы

```

# Гипотезы

1. **Гипотеза о длительных контрактах**:  
   Клиенты с долгосрочными контрактами (на один или два года) менее склонны к оттоку по сравнению с клиентами, имеющими месячные контракты.  
   **Обоснование**: Долгосрочные контракты могут предлагать более низкие цены и лучшие условия, что способствует повышению лояльности клиентов.

2. **Гипотеза о семейном статусе**:  
   Клиенты, состоящие в браке (или имеющие партнера), имеют меньшую вероятность оттока.  
   **Обоснование**: Ожидается, что клиенты с партнерами более стабильны в своих расходах и менее склонны к изменениям.

3. **Гипотеза о наличии иждивенцев**:  
   Клиенты, имеющие иждивенцев, более склонны к оттоку.  
   **Обоснование**: Клиенты с иждивенцами могут быть более чувствительными к изменениям в своих расходах и, следовательно, чаще ищут более выгодные предложения.

4. **Гипотеза о влиянии типа интернет-сервиса**:  
   Клиенты, использующие интернет-сервис "Fiber optic", имеют более высокий процент оттока по сравнению с клиентами, использующими "DSL" или не имеющими интернет-сервиса.  
   **Обоснование**: Это может указывать на то, что клиенты, выбравшие "Fiber optic", могут быть менее удовлетворены услугами или сталкиваются с проблемами, которые приводят к их уходу. Наблюдается также корреляция между месячными расходами и типом интернет-сервиса: чем выше месячные расходы, тем выше вероятность недовольства клиентов, использующих "Fiber optic".

5. **Гипотеза о влиянии месячных расходов**:  
   Чем выше месячные расходы, тем выше вероятность оттока клиентов.  
   **Обоснование**: Хотя коэффициент корреляции между оттоком и диапазонами месячных расходов составляет 0.18, что является низким значением, можно предположить, что клиенты с высокими расходами могут быть менее удовлетворены услугами и, следовательно, более склонны к уходу.

6. **Гипотеза о влиянии общей суммы расходов**:  
   Клиенты, которые потратили больше, менее склонны к оттоку.  
   **Обоснование**: Коэффициент корреляции между оттоком и диапазонами общей суммы расходов составляет -0.17, что указывает на слабую обратную зависимость. Это может свидетельствовать о том, что клиенты, которые дольше остаются с компанией и тратят больше, могут быть более лояльными.

7. **Гипотеза о влиянии срока обслуживания**:  
   Чем дольше клиент остается с компанией, тем ниже вероятность его оттока.  
   **Обоснование**: Коэффициент корреляции между оттоком и сроком обслуживания составляет -0.34, что указывает на умеренную отрицательную зависимость. Это может означать, что клиенты, которые дольше остаются, более удовлетворены услугами.

8. **Гипотеза о методах оплаты**:  
   Клиенты, использующие метод оплаты "Electronic check", имеют более высокий уровень оттока по сравнению с другими методами оплаты.  
   **Обоснование**:
   - **Удобство и доступность**: Метод "Electronic check" может восприниматься как менее удобный по сравнению с автоматическими кредитными картами, так как требует дополнительных действий со стороны клиента для подтверждения транзакций. Это может привести к ошибкам и задержкам в оплате, вызывая недовольство.
   - **Чувствительность к расходам**: Клиенты, использующие "Electronic check", могут быть более чувствительными к изменениям в ценах и условиях обслуживания. Если они сталкиваются с повышением тарифов или ухудшением качества услуг, они могут быть более склонны искать альтернативные варианты.
   - **Проблемы с платежами**: Использование "Electronic check" может быть связано с более частыми проблемами с платежами (например, недостаток средств), что может привести к задержкам в оплате и оттоку клиентов.
   - **Социально-экономический статус**: Клиенты, предпочитающие "Electronic check", могут представлять группу с более низким доходом или нестабильным финансовым положением, что делает их более подверженными оттоку в условиях экономической неопределенности.
     

9. **Гипотеза о влиянии срока обслуживания на отток клиентов**:  
   Клиенты, которые являются новыми (с коротким сроком обслуживания, особенно в первый месяц), имеют более высокий риск оттока, особенно если их месячные затраты высоки.  
   **Обоснование**:
   - **Психология нового клиента**: Новые клиенты часто оценивают услуги компании и могут быстро сравнивать их с теми, которые они получали у предыдущих провайдеров. Если они не удовлетворены, могут быстро принять решение о переходе к конкурентам.
   - **Высокие месячные затраты**: Высокие месячные расходы могут вызывать недовольство, особенно если клиенты не видят соответствующей ценности в услугах. Это недовольство может быть особенно острым в начале обслуживания.
   - **Статистическая связь**: Предполагается, что существует обратная зависимость между сроком обслуживания и вероятностью оттока. Чем меньше срок обслуживания, тем выше вероятность оттока, особенно если месячные затраты превышают определенный порог.
   - **Рынок и конкуренция**: В условиях высокой конкуренции клиенты имеют множество альтернатив и могут легко сменить провайдера, если не удовлетворены условиями. Это особенно актуально для новых клиентов, которые еще не привязались к компании.

```python
import pandas as pd
from sklearn.manifold import TSNE


# Предположим, что df_churn уже загружен и подготовлен

# Удаление столбцов, которые не нужны для модели
df_churn = df_churn.drop(columns=['Unnamed: 0', 'customerID'], errors='ignore')

# Разделение данных на признаки и целевую переменную
X = df_churn.drop(['Churn'], axis=1)  # Удаляем целевую переменную
y = df_churn['Churn']  # Целевая переменная

# Определим параметры для t-SNE
tsne_param_grid = {
    'perplexity': [5, 10, 30, 50],
    'max_iter': [250, 500, 1000]  # Изменено n_iter на max_iter
}

# Создание модели t-SNE и выбор лучших параметров
best_perplexity = None
best_n_iter = None
best_tsne_score = -1  # Инициализируем с низким значением

# Пробуем разные комбинации параметров
for perplexity in tsne_param_grid['perplexity']:
    for n_iter in tsne_param_grid['max_iter']:
        tsne = TSNE(perplexity=perplexity, max_iter=n_iter, random_state=42)
        X_tsne = tsne.fit_transform(X)  # Применяем t-SNE к полному набору данных
        
        score = np.mean(np.std(X_tsne, axis=0))  # Примерная оценка

        if score > best_tsne_score:
            best_tsne_score = score
            best_perplexity = perplexity
            best_n_iter = n_iter

print("Лучшие параметры t-SNE: perplexity =", best_perplexity, ", max_iter =", best_n_iter)

# Преобразование категориальных переменных в числовые
X = pd.get_dummies(X, drop_first=True)

# Нормализация данных
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# Применение t-SNE с лучшими параметрами
tsne = TSNE(perplexity=best_perplexity, max_iter=best_n_iter, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

```
